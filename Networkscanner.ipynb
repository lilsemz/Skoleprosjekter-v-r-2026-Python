```python
import concurrent.futures
import socket
import time
```

```python
 TARGET = "scanme.nmap.org"
```

```python
 THREADS = 100
```

```python
 PORT_RANGE = range(1,1025)
```

```python
def scan_port(target_port):
    try:
        # Opprett socket
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.settimeout(1)

        # Forsøk å koble til
        result = sock.connect_ex((TARGET, target_port))

        # Hvis det gikk bra (resultat er 0):
        if result == 0:
            print(f"[+] Port {target_port} is OPEN.")

        # Lukker så socketen
        sock.close()

    except Exception as e:
        # Gracefully handle potential errors
        print(f"[!] Error scanning port {target_port}: {e}")
```

```python
def main():
    print(f"Scanning {TARGET}...")
    print(f"Scanning ports 1-1024 with {THREADS} threads...\n")

    # Definerer start tid
    start_time = time.time()

    # Oppretter "pool" med worker threads (100 stk)
    with concurrent.futures.ThreadPoolExecutor(max_workers=THREADS) as executor:
        # Starter multithreaded scanning (her bruker vi map metoden)
        executor.map(scan_port, PORT_RANGE)

    # Definerer slutt tid
    end_time = time.time()

    # Skriver ut melding til brukeren om total tid
    print("\nScan completed in " + str(end_time - start_time) + " seconds.\n")
```

```python
 main()
```
