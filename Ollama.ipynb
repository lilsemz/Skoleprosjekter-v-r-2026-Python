```python
from ollama import chat
```

```python
prompt = "How are you today?"
messages = [
    {
        'role': 'user',
        'content': prompt,
    }
]

response = chat('llama3.2:3b', messages=messages)
print(response['message']['content'])
```

```python
tall = int(input('skriv et tall mellom 6 og 8 siffer: '))
talle = int(input('skriv et tall til med mellom 6 og 8 siffere: '))

prompt = f"calculate {tall} and add 637284. do the same with {talle} + 637284"
messages = [
    {
        'role': 'user',
        'content': prompt,
    }
]

response = chat('llama3.2:3b', messages=messages)
print(response['message']['content'])


```
