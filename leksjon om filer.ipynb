```python
total = 0.0

try:
    # Open the sales_data.txt file
    with open("sales.txt", "r") as f:
        for line in f:
            amount = float(line)
            total += amount

    # Print the total
    print(f"The total amount is: {total:,.2f}")

except FileNotFoundError:
    print("An error occurred trying to read the file")

except ValueError:
    print("An error occurred trying to read the file")

except SyntaxError:
    print("An error occurred trying to interpret the statements in the try suite")
```
