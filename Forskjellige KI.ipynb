```python
"chat"
# summer_og_hoyeste.py

# Leser inn 4 tall fra brukeren
tall1 = float(input("Skriv inn tall 1: "))
tall2 = float(input("Skriv inn tall 2: "))
tall3 = float(input("Skriv inn tall 3: "))
tall4 = float(input("Skriv inn tall 4: "))

# Legger dem i en liste
tall = [tall1, tall2, tall3, tall4]

# Regner ut sum og høyeste verdi
total_sum = sum(tall)
maks = max(tall)

# Skriver ut resultat
print(f"Høyeste tall: {maks}")
print(f"Summen av tallene: {total_sum}")
```

```python
"gemini"
import secrets

def play_random_sum_game():
    # 1. Generate a "truly" random number between 1 and 100
    # secrets.randbelow(100) gives 0-99, so we add 1
    target_number = secrets.randbelow(100) + 1

    print("--- The Truly Random Sum Challenge ---")
    print("I have chosen a secret number between 1 and 100.")

    try:
        # 2. Get user input
        num1 = float(input("Enter your first guess: "))
        num2 = float(input("Enter your second guess: "))

        user_sum = num1 + num2

        # 3. Evaluate the result
        print(f"\nYour total: {user_sum}")
        print(f"The secret number was: {target_number}")

        if user_sum == target_number:
            print("✨ Incredible! Your numbers sum up perfectly to the target.")
        else:
            difference = abs(target_number - user_sum)
            print(f"Close! You were off by {difference:.2f}.")

    except ValueError:
        print("Invalid input. Please enter numbers only.")

if __name__ == "__main__":
    play_random_sum_game()
```
